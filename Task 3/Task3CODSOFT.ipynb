AIM: Train a machine learning model that can learn from these measurements and accurately classify the Iris Flowers into their respective species  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'],categories = pd.factorize(df['species'])
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

DATA VISUALIZATION

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(df.petal_length,df.petal_width,df.species)
ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.set_zlabel('Species')
plt.title('Iris Dataset 3D PLOT')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(df.sepal_length,df.sepal_width,df.species)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Species')
plt.title('Iris Dataset 3D PLOT')
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=df)
plt.title('Iris Dataset 2D PLOT')
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=df)
plt.title('Iris Dataset 2D PLOT')
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['petal_length','petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
plt.title('Elbow Method')
plt.show()
```

APPLYING K-MEANS CLUSTERING

```python
kn = KMeans(n_clusters=3, random_state=0,)
y_pred = kn.fit_predict(df[['petal_length','petal_width']])
y_pred
```

```python
df['cluster'] = y_pred
df.head(150)
```

MEASURING THE ACCURACY

```python
from sklearn.metrics import confusion_matrix
cn = confusion_matrix(df['species'],df['cluster'])
cn
```

```python
true_label = df['species']
predicted_label = df['cluster']

cn = confusion_matrix(true_label,predicted_label)
class_label = ['setosa','versicolor','virginica']

plt.imshow(cn,interpolation='nearest',cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_label))
plt.xticks(tick_marks,class_label, rotation=45)
plt.yticks(tick_marks,class_label)

for i in range(len(class_label)):
  for j in range(len(class_label)):
    plt.text(j,i,str(cn[i,j]),ha='center',va='center',color='white')

plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
